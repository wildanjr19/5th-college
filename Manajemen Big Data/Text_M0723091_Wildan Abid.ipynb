# Data

Data yang digunakan bersumber dari Project Gutenberg, yaitu novel Alice In Wonderland (Chapter 1)

## Import Data

```python
with open('data/AliceInWonderland_Ch01.txt', 'r', encoding = 'utf-8') as f:
    docs = f.read()
```

```python
# total karakter
print("Total karakter:", len(docs))
```

# Token Awal

```python
from nltk.tokenize import word_tokenize

token_awal = word_tokenize(docs)
# banyak token awal (sebelum praproses)
print("Token awal:", len(token_awal))
```

# Praproses

## Remove Punctuation

```python
import string
```

```python
# hapus titik, koma, dan tanda baca lainnya
new_docs = docs.translate(str.maketrans('', '', string.punctuation))

print(new_docs[:5000])
```

“without pictures or conversations” dsb, tidak ikut dibersihkan karena bukan ASCII 

```python
import re
```

```python
# hapus tanda baca yang lain 
# termasuk yang bukan ASCII
new_docs = re.sub(r'[^\w\s]', '', new_docs)
print(new_docs[:5000])
```

## Stopwords

Kata yang berdiri sendiri dan tidak ada maknanya, seperti kata hubung

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# import english stopwords
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# gunakan stopwords bahasa inggris
stop_words = set(stopwords.words('english'))

# tokenisasi
word_tokens = word_tokenize(new_docs)
```

```python
# apply stopword 
filtered_docs = [word for word in word_tokens if not word.lower() in stop_words]
```

```python
# print
print(" ".join(filtered_docs[:200]))
```

Kata seperti 'was' dihapus

## Lowercase

```python
# mengubaah teks di filtered_docs menjadi lower semua
# karena bentuknya list maka gunakan list comprehension
lowered_docs = [word.lower() for word in filtered_docs]
print(lowered_docs[:5000])
```

## Stemming

Ternyata stemming tidak memberikan hasil yang baik, jadi tidak memilih stemming untuk dibagian praproses

## Lemmatization

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

wnl = WordNetLemmatizer()
```

```python
# karena terpisah perkata, gabungkan dulu
words = word_tokenize(" ".join(lowered_docs))

new_text = " ".join([wnl.lemmatize(w) for w in words])
print(new_text[:500])
```

# Tokenisasi

```python
# tokenisasi
data_tokenized = word_tokenize(" ".join(lowered_docs))
print(data_tokenized[:500])
```

## Token Akhir

```python
# melihat banyak token terakhir setelah praproses
len(data_tokenized)
```

Dari total token awal sebanyak 2615 tersisa 977 token setelah dilakukan praproses

# Text Vectorization

```python
# karena token cukup banyak, ambil 500 token pertama saja
data_tokenized = data_tokenized[:500]
print(data_tokenized)
```

## Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
# input harus berupa string, jadi gabungkan token kembali
X = vectorizer.fit_transform([" ".join(data_tokenized)])

print(X.toarray())
print(vectorizer.get_feature_names_out())
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
# input harus berupa string, jadi gabungkan token kembali
X = vectorizer.fit_transform([" ".join(data_tokenized)])

print(X.toarray())
print(vectorizer.get_feature_names_out())
```
