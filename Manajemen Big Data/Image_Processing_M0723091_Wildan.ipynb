# Normalisasi

# Noise Reduction

```python
# import gambar dalam rgb
image_full_rgb = cv2.imread('data/balinese_train_000061.jpg', cv2.COLOR_BGR2RGB)

# resize ke 320x240
image_full_rgb = cv2.resize(image_full_rgb, (240, 320))
```

```python
# tambahkan gaussian blur
gaussianblur_image = cv2.GaussianBlur(image_full_rgb, (3,3), 0)
gaussianblur_image
```

```python
# tampilkan gambar
plt.imshow(gaussianblur_image)
plt.axis('off')
plt.show()
```

# Rotation dan Translation

```python
# dapatkan dimensi
height, width = image_full_rgb.shape[:2]

# menentukan pusat rotasi
center = (width // 2, height // 2)

# menentukan sudut rotasi
angle = 45

# menentukan faktor skala
scale = 1.0

# mendapatkan matriks rotasi
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_image = cv2.warpAffine(image_full_rgb, rotation_matrix, (width, height))

# tampilkan gambars
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```

# Edge Detection

```python
# Sobel operator
sobelx = cv2.Sobel(image_full_grayscale, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobely = cv2.Sobel(image_full_grayscale, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges

# Compute gradient magnitude
gradient_magnitude = cv2.magnitude(sobelx, sobely)

# Convert to uint8
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

# Display result
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')
plt.show()
```
